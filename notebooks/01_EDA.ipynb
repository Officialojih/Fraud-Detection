## ðŸ§© Feature Distribution Analysis

In this section, we explore how the main features differ between **fraudulent** and **non-fraudulent** transactions.  
Since the data is highly imbalanced, visualizing the distributions helps us identify which features carry strong signals for detecting fraud.

```python
df['Class'].value_counts(normalize=True)
```

```python
top_features = ['V10', 'V12', 'V14', 'V17']
plt.figure(figsize=(12,8))

for i, feature in enumerate(top_features, 1):
    plt.subplot(2, 2, i)
    sns.kdeplot(data=df[df['Class']==0], x=feature, label='Non-Fraud', fill=True, alpha=0.5)
    sns.kdeplot(data=df[df['Class']==1], x=feature, label='Fraud', fill=True, alpha=0.5)
    plt.title(f"Distribution of {feature} by Class")
    plt.legend()

plt.tight_layout()
plt.show()
```

### ðŸ” Distribution of Highly Correlated Features

From earlier correlation analysis, `V10`, `V12`, `V14`, and `V17` showed strong relationships with the fraud class.  
Here we visualize how these features are distributed for both classes.

**Observations:**
- `V10` and `V14` show **significant negative values** for fraud cases, distinct from normal transactions.
- `V12` and `V17` also display **shifts in distribution**, indicating potential discriminative power.
- These patterns suggest that the PCA-transformed components still encode meaningful fraud signals.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amount by Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df[df['Class']==0], x='Hour', label='Non-Fraud', fill=True, alpha=0.5)
sns.kdeplot(data=df[df['Class']==1], x='Hour', label='Fraud', fill=True, alpha=0.5)
plt.title("Transaction Hour Distribution by Class")
plt.legend()
plt.show()
```

### ðŸ’³ Transaction Amount & â° Transaction Time Analysis

- **Transaction Amount:** Fraudulent transactions are **not necessarily high-value**; many are within normal ranges.  
  This implies that amount alone is not a strong fraud indicator.
  
- **Transaction Hour:** Fraudulent activity appears slightly more frequent at **off-peak hours** (late night or early morning).  
  This temporal pattern might help refine detection logic when combined with other features.


```python
plt.figure(figsize=(15,30))
for i, col in enumerate([f'V{i}' for i in range(1, 29)], 1):
    plt.subplot(10, 3, i)
    sns.kdeplot(data=df[df['Class']==0], x=col, fill=True, alpha=0.5)
    sns.kdeplot(data=df[df['Class']==1], x=col, fill=True, alpha=0.5)
    plt.title(col)
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```

### ðŸ“ˆ PCA Feature Distribution Overview

By visualizing all PCA features (`V1â€“V28`), we can see:
- Most features overlap significantly between fraud and non-fraud classes.
- A few components (e.g., `V10`, `V12`, `V14`, `V17`) show **distinct tails or shifts**, confirming their importance.
- This visualization helps in prioritizing features for further analysis and model interpretation.


## âœ… Summary â€” Key Insights

- Fraudulent transactions show **distinct statistical patterns** in several PCA components.
- Features `V10`, `V12`, `V14`, and `V17` emerge as strong predictors of fraudulent behavior.
- Transaction `Amount` is **not a reliable standalone indicator**, but `Time` may reveal subtle behavioral trends.
- These visual insights provide valuable intuition for feature engineering and model design in the next phase.
